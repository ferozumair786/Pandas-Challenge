```python
import pandas as pd
import numpy as np
```

```python
#import schools data from csv
schools_df = pd.read_csv('schools_complete.csv')
schools_df.head()
```

```python
#import student data from csv
students_df = pd.read_csv('students_complete.csv')
students_df.head()
```

```python
#Create new dataframe with only district schools
district_df = schools_df[schools_df['type'] == 'District']
district_df.head()
```

```python
#merge district dataframe with students dataframe to get a complete dataset
district_df = district_df.merge(students_df, how='inner', on='school_name')
district_df.head()
```

```python
#calculate fields for district summary
total_schools_district = schools_df[schools_df['type'] == 'District']['school_name'].count()
total_students_district = district_df['student_name'].count()
total_budget_district = schools_df[schools_df['type'] == 'District']['budget'].sum()
avg_math_district = district_df['math_score'].mean()
avg_read_district = district_df['reading_score'].mean()
```

```python
#create new dataframe with passing match grades and calculate the total number of students 
pass_math_district = district_df[district_df['math_score'] >= 70]['math_score'].count()
percent_pass_math = (pass_math_district/total_students_district)*100
print(percent_pass_math)
```

```python
#create new dataframe that only has passing reading score and calculate number of students
pass_read_district = district_df[district_df['reading_score'] >= 70]['reading_score'].count()
percent_pass_read = (pass_read_district/total_students_district)*100
print(percent_pass_read)
```

```python
#Overall Passing Rate (Average of the above two)
percent_pass_overall = (percent_pass_read + percent_pass_math)/2
print(percent_pass_overall)
```

```python
#enter the above data into a dictionary
Data = {'Total Schools': [total_schools_district],
       'Total Students': [total_students_district],
       'Total Budget': [total_budget_district],
       'Average Math Score': [avg_math_district],
       'Average Reading Score': [avg_read_district],
       '% Passing Math': [percent_pass_math],
       '% Passing Reading': [percent_pass_read],
       '% Overall Passing Rate': [percent_pass_overall]}
#create a new dataframe with the above dictionary
district_summary_df = pd.DataFrame (Data)
```

```python
#create a new dataframe for this part of the exercise
school_summary_df = schools_df
school_summary_df.head()
```

```python
#create new column called 'Per Student Budget' that contains budget divided by student
school_summary_df['Per Student Budget'] = school_summary_df['budget']/school_summary_df['size']
```

```python
#create new dataframe that merges our last dataframe with the original students one
ss_df = school_summary_df.merge(students_df, how='inner', on='school_name')
ss_df.head()
```

```python
#group the dataframe by school name, type, size, budget and per student budget to preserve those values
ss_avg_df = ss_df[['school_name', 'type', 'size', 'budget', 'Per Student Budget', 'reading_score', 'math_score']].groupby(['school_name', 'type', 'size', 'budget', 'Per Student Budget']).mean()
#reset the indices to columns after groupby
ss_avg_df.reset_index(level=['type', 'size', 'budget', 'Per Student Budget'], inplace=True)
ss_avg_df.head()
```

```python
#repeat similar process as earlier by getting a new dataframe for all the math passing students
ss_math_df = ss_df[(ss_df['math_score'] >= 70)]
ss_math_df = ss_math_df[['school_name','math_score']].groupby('school_name').count()
ss_math_df = ss_math_df.rename(columns={"math_score" : "total_pass_math"})
ss_math_df.head()
```

```python
#create a new dataframe with all those that passed reading
ss_read_df = ss_df[(ss_df['reading_score'] >= 70)]
ss_read_df = ss_read_df[['school_name','reading_score']].groupby('school_name').count()
ss_read_df = ss_read_df.rename(columns={"reading_score" : "total_pass_read"})
ss_read_df.head()
```

```python
 #join the above two dataframes
rm_df = ss_read_df.join(ss_math_df, how='inner')
```

```python
rm_df.head()
```

```python
#merge the above dataframe back with the average score dataframe
ss_fin_df = ss_avg_df.merge(rm_df, how='inner', on='school_name')
ss_fin_df.head()
```

```python
#calculate values from this combined table
ss_fin_df['% Passing Reading'] = (ss_fin_df['total_pass_read']/ss_fin_df['size'])*100
ss_fin_df['% Passing Math'] = (ss_fin_df['total_pass_math']/ss_fin_df['size'])*100
ss_fin_df['Overall Passing'] = (ss_fin_df['% Passing Math'] + ss_fin_df['% Passing Reading'])/2
#return schoolname to column
ss_fin_df.reset_index(level='school_name', inplace=True)
ss_fin_df.head()
```

```python

#rename columns
ss_fin_df = ss_fin_df.rename(columns={'school_name': 'School Name',
                         'type': 'School Type',
                         'size': 'Total Students',
                         'budget': 'Total Budget',
                         'reading_score': 'Avg Reading Score',
                         'math_score': 'Avg Math Score'})
ss_fin_df.head()
```

```python
#drop useless column
ss_fin_df = ss_fin_df.drop(columns=['total_pass_read', 'total_pass_math'])
```

```python
ss_fin_df.head()
```

```python
#sort the table
ss_fin_df = ss_fin_df.sort_values(by='Overall Passing', ascending=False)
```

```python
#using this syntax means it will start 5 from the last row and count down to the last row
ss_bot5 = ss_fin_df.iloc[-5:]
ss_bot5
```

```python
#calculate maximum and minimum values and step size to segment the data dynamically
minimum = int(round(ss_fin_df['Per Student Budget'].min()))
maximum = int(round(ss_fin_df['Per Student Budget'].max()))
#dividing by 2 gave the optimal step size. This is the only hardcoded value here
step = int(round((maximum-minimum)/2))
bins = []
labels = []
#creates enough bins to fit the whole dataset automatically
for i in range(minimum-step,maximum+step, step):
    bins.append(i)
#creates all the labels based on the bin above
for l in range(1,len(bins),1):    
    if l == 1:
        labels.append('< '+ str(bins[1]))
    elif l == len(bins)-1:
        labels.append('> ' + str(bins[l-1]))
    else:
        labels.append('between ' + str(bins[l-1]) + ' and ' + str(bins[l]))
```

```python
labels
```

```python
#create new columns that will sort our data into categories that we can then group by
ss_fin_df['Category'] = pd.cut(ss_fin_df['Per Student Budget'], bins, labels=labels)
```

```python
ss_fin_df.head()
```

```python
#group by category to get average value per category
ss_fin_df[['Category', 'Avg Reading Score', 'Avg Math Score', '% Passing Reading', '% Passing Math', 'Overall Passing']].groupby('Category').mean()
```

```python
#calculate maximum and minimum values and step size to segment the data dynamically
minimum = int(round(ss_fin_df['Total Students'].min()/1000))*1000
maximum = int(round(ss_fin_df['Total Students'].max()/1000))*1000
#dividing by 4 gave the optimal step size. This is the only hardcoded value here
step = int(round((maximum-minimum)/4))
bins = []
labels = []
#creates all the bins automatically
for i in range(0,maximum+step, step):
    bins.append(i)
#creates all the labels based on the bin above
for l in range(1,len(bins),1):    
    if l == 1:
        labels.append('< '+ str(bins[1]))
    elif l == len(bins)-1:
        labels.append('> ' + str(bins[l-1]))
    else:
        labels.append('between ' + str(bins[l-1]) + ' and ' + str(bins[l]))
```

```python
#sort data into bins
ss_fin_df['School Size'] = pd.cut(ss_fin_df['Total Students'], bins, labels=labels)
```

```python
ss_fin_df.head()
```

```python
#group the data by the school size to give statistics
ss_fin_df[['School Size', 'Avg Reading Score', 'Avg Math Score', '% Passing Reading', '% Passing Math', 'Overall Passing']].groupby('School Size').mean()
```

```python
#no need for bins here we already have our categories
ss_fin_df[['School Type', 'Avg Reading Score', 'Avg Math Score', '% Passing Reading', '% Passing Math', 'Overall Passing']].groupby('School Type').mean()
```
